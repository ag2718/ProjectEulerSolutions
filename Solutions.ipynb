### Problem 4 - Largest Palindrome Product

```python
def is_palindrome(num): 
    str_num = str(num) 
    if len(str_num) == 1: 
        return True 
    elif len(str_num) == 2: 
        return str_num[0] == str_num[1] 
    return str_num[0] == str_num[-1] and is_palindrome(str_num[1:-1]) 

product_list = [i * j for i in range(1000) for j in range(1000)]
product_list.sort(reverse=True)

for num in product_list:
    if is_palindrome(num):
        largest_palindrome = num
        break

largest_palindrome

```

### Problem 5 - Smallest Multiple

```python
least_common_multiple = 1

prime_factors_list = list(sp.primerange(0, 20))
prime_factor_powers = {prime_factor : 0 for prime_factor in prime_factors_list}

for prime_factor in prime_factors_list:
    for num in range(1, 20):
        if num % (prime_factor ** prime_factor_powers[prime_factor]) == 0:
            current_power = prime_factor_powers[prime_factor]
            
            while num % (prime_factor ** (prime_factor_powers[prime_factor] + 1)) == 0:
                prime_factor_powers[prime_factor] += 1
    

    least_common_multiple *= (prime_factor ** prime_factor_powers[prime_factor])
            

least_common_multiple
```

### Problem 12 - Highly Divisible Triangular Number

```python
def num_divisors(num):
    primes_list = list(sp.primerange(0, num))
    prime_factor_powers = {prime_factor : 0 for prime_factor in primes_list}

    for prime_factor in primes_list:
        while num % (prime_factor ** (prime_factor_powers[prime_factor] + 1)) == 0:
            prime_factor_powers[prime_factor] += 1 

    try:
        return reduce(lambda a, b : a * b, [i + 1 for i in list(prime_factor_powers.values())])
    except TypeError:
        return -1
```

```python
import time
triangular_num = 1

while True:
    start_time = time.time()
    if num_divisors(triangular_num * (triangular_num + 1) / 2) > 500:
        break
    triangular_num += 1
    print(time.time() - start_time)

triangular_num * (triangular_num + 1) / 2
```

### Problem 14 - Longest Collatz Sequence

```python
def collatz_length(start):
    length = 1
    num = start
    while not num == 1:
        length += 1
        num = num / 2 if num % 2 == 0 else 3 * num + 1

    return length
```

```python
collatz_length(137106)
```

### Problem 17 - Number Letter Counts

```python
def num_letters(number):

    units_digit_letter_counts = {
        0: 0,
        1: len("one"),
        2: len("two"),
        3: len("three"),
        4: len("four"),
        5: len("five"),
        6: len("six"),
        7: len("seven"),
        8: len("eight"),
        9: len("nine")
    }

    tens_digit_letter_counts = {
        0: 0,
        1: len("ten"),
        2: len("twenty"),
        3: len("thirty"),
        4: len("forty"),
        5: len("fifty"),
        6: len("sixty"),
        7: len("seventy"),
        8: len("eighty"),
        9: len("ninety")
    }

    hundreds_digit_letter_counts = {
        0: 0,
    }

    for digit in range(10):
        hundreds_digit_letter_counts[digit] = units_digit_letter_counts[digit] + len("hundred")

    if number < 10:
        units_digit, tens_digit, hundreds_digit = number, 0, 0
    elif number < 100:
        units_digit, tens_digit = [int(digit) for digit in str(number)]
        hundreds_digit = 0
    elif number < 1000:
        units_digit, tens_digit, hundreds_digit = [int(digit) for digit in str(number)]


    if number == 1000:
        return len("thousand")
    elif hundreds_digit == 0 or (tens_digit == 0 and units_digit == 0):
        return hundreds_digit_letter_counts[hundreds_digit] + tens_digit_letter_counts[tens_digit] + units_digit_letter_counts[units_digit]
    else:
        return hundreds_digit_letter_counts[hundreds_digit] + tens_digit_letter_counts[tens_digit] + units_digit_letter_counts[units_digit] + 3
```

```python
sum([num_letters(i) for i in range(1, 1001)])
```

### Problem 22 - Name Scores

```python
def calculate_name_score(name):
    letter_numbers = {letter: num for num, letter in enumerate(ascii_lowercase, start=1)}

    name_score = 0

    for letter in name.lower(): 
        name_score += letter_numbers[letter]
    
    return name_score


```

```python
with open("problem-22-names.txt") as names:
    names_list = names.read().strip().replace("\"", "").split(',')

names_list.sort()

sum([number * calculate_name_score(name) for number, name in enumerate(names_list, start=1)])
```
